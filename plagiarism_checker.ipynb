```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load a pre-trained model for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example reference dataset and input text
reference_texts = [
    "The cat sleeps on the mat every morning.",
    "The dog runs in the park with joy."
]
input_text = "The cat sleeps on the mat today."

# Generate embeddings
reference_embeddings = model.encode(reference_texts, convert_to_tensor=True)
input_embedding = model.encode([input_text], convert_to_tensor=True)

# Calculate cosine similarity
similarity_scores = cosine_similarity(input_embedding, reference_embeddings)[0]
threshold = 0.8  # Adjust threshold for plagiarism detection

# Check for plagiarism
for i, score in enumerate(similarity_scores):
    if score > threshold:
        print(f"Potential plagiarism detected! Similar to reference {i+1} with score: {score:.4f}")
    else:
        print(f"No plagiarism for reference {i+1}. Score: {score:.4f}")
```
